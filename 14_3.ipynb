```python
import pandas as pd
from scipy.stats import spearmanr
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/s-a-nersisyan/HSE_bioinformatics_2021/master/seminar14/healthy_breast.tsv', index_col = 0, sep = '\t')
df
```

```python
df['spearman'] = [spearmanr(df.loc['SPI1'], df.loc[gene])[0] for gene in df.index]
df
```

```python
df = df.loc[df['spearman'].abs() > 0.8]
```
