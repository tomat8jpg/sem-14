```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/s-a-nersisyan/HSE_bioinformatics_2021/master/seminar14/BRCA_pam50.tsv', index_col = 0, sep = '\t')
df
```

```python
plt.tight_layout()
esr1 = sns.boxplot(x = 'Subtype', y = 'ESR1', data = df)
esr1.set_xticklabels(esr1.get_xticklabels(), rotation=45)
```

```python
plt.tight_layout()
PGR = sns.boxplot(x = 'Subtype', y = 'PGR', data = df)
PGR.set_xticklabels(PGR.get_xticklabels(), rotation=45)
```

```python
plt.tight_layout()
ERBB2 = sns.boxplot(x = 'Subtype', y = 'ERBB2', data = df)
ERBB2.set_xticklabels(ERBB2.get_xticklabels(), rotation=45)
```

```python
plt.tight_layout()
MKI67 = sns.boxplot(x = 'Subtype', y = 'MKI67', data = df)
MKI67.set_xticklabels(MKI67.get_xticklabels(), rotation=45)
```

Normal-like экспрессирует все 4 гена, экспрессия PGR низкая
Luminal A экспрессирует все 4 гена, экспрессия ESR1 завышена
Luminal B экспрессирует все 4 гена, экспрессия ESR1 и MKI67 выше чем у Luminal A, PGR ниже, но все равно достаточно высокая
Triple-negative не экспрессирует PGR, экспрессия ESR1 почти нулевая, экспрессирует ERBB2, и самая высокая экспрессия MKI67
HER2-enriched не экспрессирует PGR, низкая экспрессия ESR1, самая высокая экспрессия ERBB2
Healthy нормально экспрессируют  ESR1, PGR и ERBB2, и очень слабо экспрессируют MKI67
